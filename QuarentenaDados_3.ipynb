```python
imdb_usa["lucro"] = imdb_usa["gross"] - imdb_usa["budget"]
budget_gross = imdb_usa.query("budget >0 | gross > 0")[["budget", "lucro"]].dropna()
sns.scatterplot(data = budget_gross, x="budget", y="lucro")
```
