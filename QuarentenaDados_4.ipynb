```python
respostas_gabarito = respostas_gabarito.dropna(subset=['TX_RESPOSTAS_CN',
                                                       'TX_RESPOSTAS_CH',
                                                       'TX_RESPOSTAS_LC',
                                                       'TX_RESPOSTAS_MT',
                                                       'TX_GABARITO_CN',
                                                       'TX_GABARITO_CH',
                                                       'TX_GABARITO_LC',
                                                       'TX_GABARITO_MT'])
```

```python
respostas_gabarito
```

```python
def calculo_acertos(respostas,gabarito):
  return [sum(c==d for c, d in zip(a,b)) for a, b in zip(respostas, gabarito)]

categorias=["CN","CH","LC","MT"]
for c in categorias:
  respostas_gabarito[f"acertos{c}"] = calculo_acertos(respostas_gabarito[f"TX_RESPOSTAS_{c}"], respostas_gabarito[f"TX_GABARITO_{c}"])

for c in categorias:
  respostas_gabarito[f'erros{c}'] = 45 - respostas_gabarito[f"acertos{c}"]

```
